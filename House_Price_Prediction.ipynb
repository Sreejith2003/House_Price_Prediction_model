```python
# Dropping the meadian house value form the test_data

X_test, y_test = test_data.drop(['median_house_value'], axis = 1), test_data['median_house_value']
```

```python
# Standardizing the data

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train_s = scaler.fit_transform(X_train)
X_test_s = scaler.transform(X_test)
```

```python
forest = RandomForestRegressor()
forest.fit(X_train_s, y_train)
```

```python
forest.score(X_test_s, y_test)
```

```python
# The Score is low so we approach another method as Grid Search by using random forest
```

```python
# Optimizing Hyperparameter Tuning

from sklearn.model_selection import GridSearchCV

forest = RandomForestRegressor()

para_grid = {
    "n_estimators" : [30, 50, 100],
    "min_samples_split" : [2,4],
    "max_depth": [None,4,8]
}

grid_search = GridSearchCV(forest, para_grid, cv = 5,
                           scoring = 'neg_mean_squared_error', return_train_score = True)

grid_search.fit(X_train_s, y_train)
```

```python
grid_search.best_estimator_
```

```python
grid_search.best_estimator_.score(X_test_s, y_test)
```

```python
# Gird search also does not provide accurate result so we approach Cross Validation
```

```python
# Cross validation the Score

from sklearn.model_selection import cross_val_score

scores = cross_val_score(forest, X_train_s, y_train, cv=5)
print(f'Cross-Validation Scores: {scores}')
print()
print(f'Average CV Score: {scores.mean()}')
```
